## SELECT

```sql
WHERE [condicionales]
GROUP BY [agrupamiento por columnas]
ORDER BY [ordenamiento a partir de columnas]
LIMIT [número máximo de registros]
```

La cláusula `WHERE` sirve como condicional en las queries de sql. Es importante tener presente los operadores relacionales:

|Operador|Significado|
|---|---
|=|igual a|
|<>|distinto de|
|<|menor que|
|>|mayor que|
|<=|menor o igual que|
|>=|mayor o igual que|

Adicional a esto, si se va a utilizar más de un condicional, es necesario emplear operadores lógicos:

|Operador|Operación|
|---|---|
|AND|conjunción|
|OR|disyunción|
|NOT|negación|

Si se debe hacer alguna operación a partir de un valor, se puede emplear la cláusula `CASE`.

```sql
CASE
  WHEN [primera condición]
    THEN [valor o cálculo a partir de la primera condición]
  WHEN [segunda condición]
    THEN [valor o cálculo a partir de la segunda condición]
  ELSE [valor o cálculo si ninguna se cumple]
END

```

### JOINS

```sql
SELECT [columnas a devolver]
FROM [tabla de la izquierda]
[JOIN TYPE] [tabla de la derecha]
on [tabla de la izquiera].[variable en izquierda para hacer match] = [tabla de la derecha].[variable en derecha para hacer match]
```
#### Tipos de JOINS
```sql
LEFT JOIN
RIGHT JOIN
INNER JOIN
```
