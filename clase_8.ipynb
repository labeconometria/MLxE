# LÃ­nea de regresiÃ³n

```python
np.random.seed(100)

x = np.arange(0,10,0.1)
e = np.random.normal(0,4,100)

y = 5 + 4*x + e
```

```python
plt.scatter(x,y)
```

```python
fit_1 = np.polyfit(x,y,1)
fit_2 = np.polyfit(x,y,2)
```

```python
fit_1
```

```python
fit_2
```

```python
y_pred_1 = fit_1[1] + fit_1[0] * x
y_pred_2 = fit_2[2] + fit_2[1] * x + fit_2[0] * (x**2)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.plot(x, y_pred_1, color='r', label='Grado 1')
plt.plot(x, y_pred_2, color='g', label='Grado 2')
plt.legend()
plt.show()
```

### Line styles

-`solid` o `-`

-`dashed` o `--`

-`dashdot` o `-.`

-`:`

```python
y = [np.sin(i) for i in np.arange(0,10,0.1)]

line_styles = ['-','--','-.',':']

for i,x in enumerate(line_styles):
  plt.plot([n*(i+1) for n in y],ls=x, label=x)

plt.legend()
plt.show()
```

## Mean-and-error plots

```python
x = np.arange(0, 5, 0.3)
y = np.sin(-x)

np.random.seed(100)
e = 0.1 * np.abs(np.random.randn(len(y)))

plt.errorbar(x,y, yerr=e, fmt='.-')
plt.show()
```
